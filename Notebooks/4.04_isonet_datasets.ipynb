```python
data_path = os.path.join(PARENT_PATH, 'data/isoNet/cryoCARE_dataset/cryoCARE_results/data/')
train_x = glob(data_path+'train_x/*')
train_y = glob(data_path+'train_y/*')

len(train_x), len(train_y)
```

```python
test_x = glob(data_path+'test_x/*')
test_y = glob(data_path+'test_y/*')

len(test_x), len(test_y)
```

```python
counter = 0
for x, y in zip(sorted(train_x), sorted(train_y)):
    if counter%100==0:
        x_arr = read_array(x)
        y_arr = read_array(y)
        fig, ax = plt.subplots(1, 2, figsize=(12, 10))
        z_idx = x_arr.shape[0]//2
        ax[0].imshow(x_arr[z_idx])
        ax[1].imshow(y_arr[z_idx])
        plt.show()
    counter+=1
```
